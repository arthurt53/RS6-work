```python
import os
import pandas as pd
import time 
from tqdm import tqdm
#将爬取的数据合并成单一数据集
filePath = 'D:\Jupyter notebook\learning\l13\data'
file_list = os.listdir(filePath)
temp = pd.read_csv('./data/all_data_10.csv')
col_name = temp.columns.values.tolist()
all_data = pd.DataFrame(columns = col_name)
for file in tqdm(file_list):
    data_temp = pd.read_csv('./data/'+ file)
    all_data = pd.concat([all_data,data_temp],ignore_index = True)
    
print(all_data)
```
