```python
#抓取一线城市房价
browser = webdriver.Chrome()
url_1 = "https://www.anjuke.com/fangjia/"
browser.get(url_1)
country_click = browser.find_element(By.XPATH,"/html/body/div[2]/div[2]/div/span[2]/a[1]")
country_click.click()
last_url1 = browser.current_url
first_list = browser.find_element(By.XPATH,"/html/body/div[2]/div[2]/div/span[2]/div").text.split()
first_list = first_list[1:]
first_list_1 = list()
for item_1 in first_list:
    if len(item_1) == 1:
        continue
    else:
        first_list_1.append(item_1)
#print(first_list_1)
all_data = pd.DataFrame(columns = ['省市','区域','时间','价格'])
for num_1 in range(1,len(first_list_1)):
#for num_1 in range(1,2):
    browser.get(last_url1)
    country_one_click = browser.find_element(By.XPATH,"/html/body/div[2]/div[2]/div/span[2]/div/a[" + str(num_1) + "]")
    country_one_click.click()
    last_url2 = browser.current_url
    area_list = browser.find_element(By.XPATH,"/html/body/div[2]/div[2]/div/span[2]").text.split()
    area_list = area_list[1:]
    #print(area_list)
    for num_2 in range(1,len(area_list)+1):
    #for num_2 in range(1,2):
        browser.get(last_url2)
        area_click = browser.find_element(By.XPATH,"/html/body/div[2]/div[2]/div/span[2]/a[" + str(num_2) +"]")
        area_click.click()
        area_year_num = browser.find_element(By.XPATH,"/html/body/div[2]/div[3]/div[1]").text.split()
        count_num = 0
        for item_2 in area_year_num:
            if item_2 == '更多>':
                count_num += 1
        area_address = list()
        for num_3 in range(1,count_num+1):
            area_one = browser.find_element(By.XPATH,"/html/body/div[2]/div[3]/div[1]/div["+str(num_3)+"]/h3/a")
            area_address.append(area_one.get_attribute('href'))
        for url_2 in area_address:
            browser.get(url_2)
            areaprice_list = browser.find_element(By.XPATH,"/html/body/div[2]/div[5]/div[1]/div[1]/ul").text.split()
            data_list = list()
            index_num  = int(len(areaprice_list)/3)
            for i in range(index_num):
                data_temp = list([first_list_1[num_1-1],area_list[num_2-1]])
                data_temp.append(areaprice_list[3*i+0])
                data_temp.append(areaprice_list[3*i+1])
                data_list.append(data_temp)            
            data_1 = pd.DataFrame(data_list,columns = ['省市','区域','时间','价格'])
            all_data = pd.concat([all_data,data_1], axis=0,ignore_index = True)

print(all_data)
browser.close()
```
