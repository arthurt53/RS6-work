```python
import pandas as pd
data = pd.read_csv('data_display.csv')
print(data)
```

```python
data['省市'] = data['标签'].map(lambda x:x.split('/')[0])
data['城市'] = data['标签'].map(lambda x:x.split('/')[1])
data['区域'] = data['标签'].map(lambda x:x.split('/')[2])
data_1 = data[['省市','城市','区域','时间','价格']]
print(data_1)
```

```python
data_1.to_csv('data_powerbi.csv',index=0)
```

```python
import pandas as pd
data_2 = pd.read_csv('data_powerbi.csv')
print(data_2)
```

```python
data_2.to_csv('data_powerbi_encode.csv',index=0, encoding = "gbk")
```

```python
old = '湘西'
new = '湘西土家族苗族自治州'
data_2['城市']=data_2['城市'].replace(old,new)
#print(data_2[data_2.城市 == new])
```

```python
old_name = '林芝'
new_name = old_name + '市'
data_2['城市']=data_2['城市'].replace(old_name,new_name)
#print(data_2[data_2.城市 == new_name])
```

```python
print(data_2[data_2.城市 == '恩施'])
```

```python
data_3 = data_2[['省市','时间','价格']]
data_4 = data_3.groupby(['省市','时间']).mean()
data_4 = data_4.reset_index()
print(data_4)
```

```python
data_4['价格'] = data_4['价格'].astype('int32')
```

```python
print(data_4)
```

```python
data_4.to_csv('country_mean.csv',index = 0,encoding = "gbk")
```

```python
data_5 = data_4.groupby('省市').max()
data_6 = data_4.groupby('省市').min()
data_7 = data_4[data_4.时间 == '2020-04-01']
data_5 = data_5.reset_index(col_fill='省市')
data_6 = data_6.reset_index(col_fill='省市')
data_5 = data_5.drop('时间',axis = 1)
data_6 = data_6.drop('时间',axis = 1)
data_7 = data_7.drop('时间',axis = 1)
data_5 = data_5.rename({'价格':'最大值'})
data_6 = data_6.rename({'价格':'最小值'})
data_7 = data_7.rename({'价格':'当前值'})
print(data_5,data_6,data_7)
```

```python
data_8 = data_5.merge(data_6,on = '省市',how ='outer')
print(data_8)
```

```python
data_8 = data_8.merge(data_7,on = '省市',how ='outer')
```

```python
data_8 = data_8.rename({'价格_x':'最大值','价格_y':'最小值','价格':'当前价'},axis='columns')
print(data_8)
```

```python
data_8.to_csv('country_price_pbi.csv',index = 0 ,encoding = 'gbk')
```
